```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
asthma1 = pd.read_csv('data/asthma_adults.csv')
```

```python
asthma1
```

```python
asthma = asthma1.drop(columns = ['|| \n||', 'Standard \nError', 'Column9','95% CId \n(Percent)', 'Weighted \nNumbere', '95% CId \n(Weighted Number)'])
asthma
```

```python
asthma = asthma.drop([0,51,52,53,54])
```

```python
asthma
```

```python
asthma = asthma.reset_index(drop=True)
asthma
```

```python
asthma.columns=['state', 'sample_size','prevalence_perc' ]
```

```python
asthma
```

```python
asthma.info()
```

```python
asthma.prevalence_perc = asthma.prevalence_perc.astype(float)
```

```python
plt.figure(figsize=(16,10))
colors = sns.color_palette('coolwarm', len(asthma))
sns.barplot(x='state', y='prevalence_perc', data=asthma, palette=colors)
for index, value in enumerate(asthma['prevalence_perc']):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')
plt.title('Asthma Prevalence by State')
plt.xlabel('State')
plt.ylabel('Prevalence Percentage')
plt.grid(True, axis='y')
plt.xticks(rotation=45)          
plt.tight_layout()
plt.show();
```

```python
asthma_top10 = asthma.nlargest(10,'prevalence_perc')
asthma_top10.reset_index(drop=True)
```

```python
sns.barplot(x='state', y='prevalence_perc',data=asthma_top10) 
plt.title('TOP 10 STATES ASTHMA REPORTED')
plt.show();
```

```python
asthma_adults = pd.read_csv('data/asthma_adults_2021.csv')
```

```python
asthma_adults.columns =['state', 'asthma_total', 'asthma_perc']
asthma_adults.asthma_perc = asthma_adults.asthma_perc.str[:-7]
asthma_adults
```

```python
asthma_adults.info()
```

```python
asthma_adults = asthma_adults.drop([9,51,52,53])
asthma_adults = asthma_adults.reset_index(drop=True)
```

```python
asthma_adults.asthma_perc = asthma_adults.asthma_perc.astype(float)
```

```python
asthma_adults.info()
```

```python
plt.figure(figsize=(16,10))
colors = sns.color_palette('coolwarm', len(asthma_adults))
sns.barplot(x='state', y='asthma_perc', data=asthma_adults, palette=colors)
for index, value in enumerate(asthma_adults['asthma_perc']):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')
plt.title('Asthma Prevalence In Adults by State')
plt.xlabel('State')
plt.ylabel('Prevalence Percentage')
plt.grid(True, axis='y')
plt.xticks(rotation=45)          
plt.tight_layout()
plt.show();
```

```python
asthma_top10_adults = asthma_adults.nlargest(10,'asthma_perc')
```

```python
asthma_top10_adults = asthma_top10_adults.reset_index(drop=True)
asthma_top10_adults
```

```python
sns.barplot(x='state', y='asthma_perc',data=asthma_top10_adults) 
plt.title('TOP 10 STATES ASTHMA REPORTED IN ADULTS')
plt.xticks(rotation=45)  # Adjust the rotation angle as needed
plt.show();
```
